```python
def write_cases(algorithm, device_type):
    sizes = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
    for size in sizes:
        size_a = size;
        size_b = size;
        print(f"algorithm={algorithm} device_type={device_type} size_a={size_a} size_b={size_b} iterations={25}")
```

```python
import math
```

```python
for m in range(1, 9):
    for n in range(1, 9):
        write_cases(f"tiled_st_{8}_{m}_{n}", "cpu")
        write_cases(f"tiled_st_{16}_{m}_{n}", "cpu")
```

```python
# evenly distributed numbers
def write_cases_evenly(algorithm, device_type):
    mn = 1000
    mx = 10_000_000_000
    steps = 100
    for step in range(steps + 1):
        size2 = mn + step * (mx - mn) / steps
        size = int(math.sqrt(size2))
        # print(int(math.sqrt(size2)))
        size_a = size
        size_b = size
        print(f"algorithm={algorithm} device_type={device_type} size_a={size_a} size_b={size_b} iterations={2}")
```

```python
write_cases_evenly(f"tiled_mt_16_4_6_4", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_8", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_16", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_32", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_64", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_128", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_256", "cpu")
write_cases_evenly(f"tiled_mt_16_4_6_512", "cpu")
```
